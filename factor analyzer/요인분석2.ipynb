```python
import pandas as pd
import numpy as np
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity, calculate_kmo
```

```python
df = pd.read_csv('final_animal_visit_travel.csv', encoding = 'cp949')
```

```python
df
```

```python
df[df['시군구'] == '울릉군']
```

```python
df2 = df[df['시군구'] != '성북구']
```

```python
df3 = df2.drop('시군구', axis=1).values
```

```python
df4 = (df3 - np.mean(df3, axis = 0))/np.std(df3, axis = 0, ddof =True)
```

```python
df4
```

```python
_, p_value = calculate_bartlett_sphericity(df4)
_, kmo_measure = calculate_kmo(df4)
```

```python
p_value, kmo_measure
```

```python
fa = FactorAnalyzer(n_factors=2, rotation = 'varimax').fit(df4)
```
