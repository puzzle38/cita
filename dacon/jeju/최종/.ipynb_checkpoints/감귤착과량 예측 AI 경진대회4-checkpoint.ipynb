```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
matplotlib.rcParams['font.family'] ='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] =False
```

```python
# 데이터 불러오기
train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')
```

```python
# train 데이터 보기
train.head()
```

```python
# test 데이터 보기
test.head()
```

```python
train['9월새순평균'] = train.filter(regex='새순').filter(regex='09-').mean(axis=1)
train['10월새순평균'] = train.filter(regex='새순').filter(regex='10-').mean(axis=1)
train['11월새순평균'] = train.filter(regex='새순').filter(regex='11-').mean(axis=1)
```

```python
train['9월엽록소평균'] = train.filter(regex='엽록소').filter(regex='09-').mean(axis=1)
train['10월엽록소평균'] = train.filter(regex='엽록소').filter(regex='10-').mean(axis=1)
train['11월엽록소평균'] = train.filter(regex='엽록소').filter(regex='11-').mean(axis=1)
```

```python
data = (train[['착과량(int)','수고(m)', '수관폭1(min)', '수관폭2(max)', '수관폭평균','9월새순평균','10월새순평균','11월새순평균','9월엽록소평균']])
data.corr()
```

```python
# 착과량과 수고,수관폭의 상관관계
data = (train[['착과량(int)','수고(m)', '수관폭1(min)', '수관폭2(max)', '수관폭평균']])
colormap = plt.cm.PuBu
plt.figure(figsize=(10, 8))
plt.title("착과량(int)과의 상관관계", y = 1.05, size = 15)
sns.heatmap(data.astype(float).corr(), linewidths = 0.1, vmax = 1.0, 
            square = True, cmap = colormap, linecolor = "white", annot = True, annot_kws = {"size" : 16})
```

```python
# train, validaton data 만들기
X_train = train[['9월새순평균','10월새순평균','11월새순평균','수고(m)', '수관폭1(min)', '수관폭2(max)', '수관폭평균']]
y_train = train['착과량(int)']
x_train, x_val, y_train, y_val = train_test_split(X_train, y_train, random_state=38)
```

```python
# 파라미터 최적화(GridSearchCV)
params = {
    'n_estimators' : (10, 300),
    'max_depth' : (4, 8),
    'min_samples_leaf' : (8, 18),
    'min_samples_split' : (8,16)
}

model = RandomForestRegressor(random_state = 38, n_jobs=-1)
grid_cv = GridSearchCV(model, param_grid=params, cv=2, n_jobs=-1)
grid_cv.fit(x_train, y_train)

print('최적 하이퍼 파라미터 :', grid_cv.best_params_)
print('최적 예측 정확도:{0:.4f}'.format(grid_cv.best_score_))
```

```python
import optuna
```

```python
def objective(trial):
    rf_max_depth = trial.suggest_int('rf_max_depth', 2, 12)
    rf_n_estimators = trial.suggest_int('n_estimators', 10, 500)
    rf_min_samples_leaf = trial.suggest_int('min_samples_leaf', 2, 18)
    rf_min_samples_split = trial.suggest_int('min_samples_split',2, 20)
    rf_random_state = trial.suggest_int('random_state',1,1000)
    
    model = RandomForestRegressor(
        random_state = rf_random_state,
        n_estimators=rf_n_estimators,
        max_depth=rf_max_depth,
        min_samples_leaf=rf_min_samples_leaf,
        min_samples_split=rf_min_samples_split
    )
    
    
    model.fit(x_train, y_train)
    pred = model.predict(x_val)
    nmae = NMAE(y_val, pred)
    return nmae
```

```python
study_rf = optuna.create_study(direction='minimize')
```

```python
study_rf.optimize(objective, n_trials=100)
```

```python
trial = study_rf.best_trial
trial_params = trial.params
print('NMAE : {},\n params {}'.format(trial.value, trial_params))
```

```python
# 최적 하이퍼 파라미터로 모델 생성
model = RandomForestRegressor(random_state=88, max_depth=5, min_samples_leaf=3, min_samples_split=13, n_estimators = 10)
model.fit(x_train, y_train)
```

```python
pred_val = model.predict(x_val)
```

```python
def NMAE(true, pred):
    mae = np.mean(np.abs(true-pred))
    score = mae / np.mean(np.abs(true))
    return score
NMAE(y_val, pred_val)
```

```python
test['9월새순평균'] = test.filter(regex='새순').filter(regex='09-').mean(axis=1)
test['10월새순평균'] = test.filter(regex='새순').filter(regex='10-').mean(axis=1)
test['11월새순평균'] = test.filter(regex='새순').filter(regex='11-').mean(axis=1)
```

```python
# 제출파일 생성
pred = model.predict(test[['9월새순평균','10월새순평균','11월새순평균']])
sample_submission = pd.read_csv('./sample_submission.csv')
sample_submission['착과량(int)'] = pred
sample_submission.to_csv('./submit_9_10_11.csv', index=False)
```
