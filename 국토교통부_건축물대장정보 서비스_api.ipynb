```python
# 라이브러리 불러오기
import requests
from bs4 import BeautifulSoup
import pandas as pd
import pprint
import bs4
```

```python
# 요청변수 지정
url = 'http://apis.data.go.kr/1613000/BldRgstService_v2/getBrTitleInfo?'
sigungu = 'sigunguCd=44133'
bjdong = '&bjdongCd=10300'
numofrows = '&numOfRows=1000'
servicekey = '&ServiceKey=UJTS9gkwTNmntwhu0ZAqujfqoUAk%2B%2Byzanzu4PXR0%2BjFgpjjwa484p4FNbK10R6dNdyt7DQ1HeuyPt9oUc2cAA%3D%3D'
```

```python
# url 받아오기
result = requests.get(url+sigungu+bjdong+numofrows+"&pageNo=1"+ servicekey)
```

```python
content = result.text
```

```python
pp = pprint.PrettyPrinter(indent=4)
```

```python
soup = bs4.BeautifulSoup(content, 'lxml-xml')
rows = soup.findAll("item")
```

```python
row_list = []
name_list = []
value_list = []

for i in range(0, len(rows)):
    columns = rows[i].find_all()
    for j in range(0, len(columns)):
        if i == 0:
            name_list.append(columns[j].name)
        value_list.append(columns[j].text)
    row_list.append(value_list)
    value_list = []
```

```python
df = pd.DataFrame(row_list, columns=name_list)
df
```

```python
df.to_csv('건축물대장_표제부_백석동.csv', encoding='cp949', index = False)
```
